```python
from langchain_ollama import ChatOllama
from langgraph.checkpoint.memory import MemorySaver
from langgraph.prebuilt import create_react_agent
from datetime import datetime
import os


llm = ChatOllama(model="gpt-oss:20b", validate_model_on_init=True)
```

```python
from langchain_mcp_adapters.client import MultiServerMCPClient
from langgraph.prebuilt import create_react_agent

client = MultiServerMCPClient(
    {
        "ensemble": {
            "command": "python",
            # Make sure to update to the full absolute path to your math_server.py file
            "args": ["/Users/hari/projects/ensemble_launcher/ensemble_launcher/tests/start_mcp.py"],
            "transport": "stdio",
        }
    }
)
tools = await client.get_tools()
```

```python
agent = create_react_agent(llm,tools)
agent_response = await agent.ainvoke({"messages": "Run CFD simulations for a range of temperature 300 to 500 K and pressure 1 to 10 atm and get the result"})
```

```python
for msg in agent_response['messages']:
    msg.pretty_print()
```
